```python
# Line Graph
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
# Adjusting the axes
```

```python
axes = plt.axes()
axes.set_xlim([2,4])
axes.set_ylim([2,4])
plt.plot(x,y)
plt.show()
```

```python
# Getting grid
```

```python
axes = plt.axes()
axes.set_xlim([2,4])
axes.set_ylim([2,4])
axes.grid()
plt.plot(x,y)
plt.show()
```

```python
# Saving a figure to local
```

```python
axes = plt.axes()
axes.set_xlim([2,4])
axes.set_ylim([2,4])
axes.grid()
plt.plot(x,y)
plt.show()
plt.savefig('C:\Users\anshuman.kundu1\Desktop\30-Days-ML-DL-AI',format='png')
```

```python
# Change line type and colors and adding labels
```

```python
X = [5,4,3,2,1]
Y = [1,2,3,4,5]
plt.xlabel('Something')
plt.ylabel('Anything')
plt.plot(x,y,'b-')
plt.plot(X,Y, 'r:')
plt.legend(['Anything', 'Something'])
plt.show()
```

```python
# Pie charts
```

```python
values = [12, 55, 4, 32, 74]
colors = ['r', 'b', 'g', 'y', 'c']
explode = [0, 0, 0.55, 0, 0.66]
labels = ['America', 'Australia', 'Russia', 'Pakistan', 'India']
plt.pie(values, explode, labels, colors)
plt.title('Countries')
plt.show()
```
